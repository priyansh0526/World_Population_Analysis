### 12.Extraction of top 10 countries with respect to total population

```python
# Filter data for total population
total_population_data = df[df["Indicator Code"] == "SP.POP.TOTL"]

# Sort Data based on the total population for 2024
total_population_sorted = total_population_data.sort_values(by="2024", ascending=False)

# Get the top 10 countries with the highest population for 2024
total_top_10_countries = total_population_sorted.head(10)
print("Top 10 countries of total population\n")
print(total_top_10_countries[['Country Name']])
```

### 13.Bar Plot

### 14.Scatter Bar Plot

#### 14.1.Top 10 Countries with Male and Female Populations (2024)


```python
# Merge male and female population data on 'Country Name'
merged_data = pd.merge(male_population_data, female_population_data, on="Country Name", suffixes=("_male", "_female"))
merged_data
```

```python
# Calculate the total population for each country (male + female)
merged_data["Total Population"] = merged_data["2024_male"] + merged_data["2024_female"]
```

```python
merged_data.head()
```

```python
# Sort data based on total population in descending order
sorted_data = merged_data.sort_values(by="Total Population", ascending=False)

# Select the top 10 countries with the highest total population
top_10_countries = sorted_data.head(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country Name", y="2024_female", data=top_10_countries, color="red", label="Female Population")
sns.barplot(x="Country Name", y="2024_male", data=top_10_countries, bottom=top_10_countries["2024_female"], color="green", label="Male Population")
plt.title("Top 10 Countries with Male and Female Populations (2024)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show() 
```

#### 14.2.Bottom 10 Countries with Male and Female Populations (2024)


```python
# Select the top 10 countries with the highest total population
bottom_10_countries = sorted_data.tail(10)
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country Name", y="2024_female", data=bottom_10_countries, color="red", label="Female Population")
sns.barplot(x="Country Name", y="2024_male", data=bottom_10_countries, bottom=bottom_10_countries["2024_female"], color="green", label="Male Population")
plt.title("Bottom 10 Countries with Male and Female Populations (2024)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
